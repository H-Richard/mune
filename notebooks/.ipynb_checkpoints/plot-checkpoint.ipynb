```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import pandas as pd
from datetime import datetime, date, timedelta
import yfinance as yf
DATE_FORMAT = "%Y-%m-%d"
TODAY = date.today()
SIXTY_DAYS_AGO = TODAY - timedelta(days=60)
index_data = yf.download('^GSPC', SIXTY_DAYS_AGO.strftime(DATE_FORMAT), TODAY.strftime(DATE_FORMAT), interval="60m").reset_index()
```

```python
index_data.tail()
```

```python
data =  yf.download('GOOG', SIXTY_DAYS_AGO.strftime(DATE_FORMAT), TODAY.strftime(DATE_FORMAT), interval="60m").reset_index()

fig = make_subplots(specs=[[{"secondary_y": True}]])
fig.add_trace(go.Candlestick(
    x = data['Datetime'],
    open = data['Open'], high = data['High'],
        low = data['Low'], close=data['Close']), secondary_y = True)
fig.add_trace(go.Bar(x = data['Datetime'], y=data['Volume']), secondary_y=False)
fig.layout.yaxis2.showgrid=False


## Removing gaps

dt_all = pd.date_range(start=data['Datetime'].iloc[0],end=data['Datetime'].iloc[-1])
dt_obs = [d.strftime("%Y-%m-%d") for d in pd.to_datetime(data['Datetime'])]
dt_breaks = [d for d in dt_all.strftime("%Y-%m-%d").tolist() if not d in dt_obs]
fig.update_xaxes(rangebreaks=[dict(values=dt_breaks)])
fig.show()
```

```python
ticker_data = yf.download('GOOG', SIXTY_DAYS_AGO.strftime(DATE_FORMAT), TODAY.strftime(DATE_FORMAT), interval="60m").reset_index()
```

```python
assert ticker_data.size == index_data.size
assert type(ticker_data['Datetime'][0]) == type(index_data['Datetime'][0])
```

```python
adjusted_data = ticker_data.join(index_data, lsuffix="_ticker", rsuffix="_index")
adjusted_data['Open_adjusted'] = adjusted_data['Open_ticker'] / adjusted_data['Open_index']
adjusted_data['Open_adjusted']
```
